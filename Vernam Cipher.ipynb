1. Write Python function to implement encryption and decryption algorithms of Vernam Cipher


```python
def randomkey(m, key):
    rkey=[]
    n=len(m)
    b1=int(key[0])
    b2=int(key[1])
    b3=int(key[2])
    for i in range(n):
        if i % 4 == 0:
            r=b1 ^ b3
        else:
            r=b2 ^ (int(key[i % 4]))
        rkey.append(r)
    return rkey

def encrypt(m, rkey):
    ctext=[]
    for i in range(len(m)):
      ctext.append(m[i] ^ rkey[i])
    return ctext

def decrypt(ctext, rkey):
    ptext=[]
    for i in range(len(ctext)):
       ptext.append(ctext[i] ^ rkey[i])
    return ptext

message=input("Enter the Plaintext: ")
message=[int(x) for x in message]
key=input("Key: ")
rkey=randomkey(message,key)
print("Random Key Generated: ",rkey)

ciphertext=encrypt(message,rkey)
plaintext=decrypt(ciphertext,rkey)

print("Ciphertext: ",ciphertext)
print("Plaintext: ",plaintext)
```
